```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("/kaggle/input/crop-production-data-india/Crop Production data.csv")
```

```python
df["Production"]=df["Production"].fillna(0)
```

```python
df.sample()
```

```python
df.isnull().sum()
```

```python
one_hot=OneHotEncoder(sparse_output=False)
one=one_hot.fit_transform(df[["State_Name","District_Name","Season","Crop"]])
one=pd.DataFrame(one,columns=one_hot.get_feature_names_out())
one["Crop_Year"]=df["Crop_Year"]
one["Area"]=df["Area"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(one,df["Production"],test_size=0.2,random_state=42)
```

```python
random_reg=RandomForestRegressor()
random_reg.fit(X_train,Y_train)
random_reg.score(X_test,Y_test)
```

```python
lin_reg=LinearRegression()
lin_reg.fit(X_train,Y_train)
lin_reg.score(X_test,Y_test)
```
