```python
import re
class name_detect:
    def __init__(self,name,dataset):
        self.name=name
        self.dataset=dataset
        self.name_find()
    def name_find(self):
        if re.search(self.name,self.dataset):
            print(f" Name is{self.name}")
        else:
            print(f"Not Found {self.name}")
```
